# Query Database

```python
import sqlite3

import pandas as pd
from tabulate import tabulate
```

```python
# Connect to the database
conn = sqlite3.connect("docs/database/recipes.db")
cursor = conn.cursor()
```

## Recipes

```python
# Query all recipes
cursor.execute("SELECT * FROM recipes")
recipes = cursor.fetchall()
```

```python
# Get column names from cursor description
column_names = [desc[0] for desc in cursor.description]
```

```python
recipes_df = pd.DataFrame(recipes, columns=column_names)

recipes_df
```

```python
# Print as a nice formatted table
print(tabulate(recipes, headers=column_names, tablefmt="pretty"))
```

## Ingredients

```python
cursor.execute("SELECT * FROM ingredients")
ingredients = cursor.fetchall()

column_names = [desc[0] for desc in cursor.description]

print(tabulate(ingredients, headers=column_names, tablefmt="pretty"))
```

```python
# Close the connection
conn.close()
```
